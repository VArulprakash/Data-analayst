**Statistical analysis for numerical columns**

```python
# MEASURE OF CENTRAL TENDENCY
numeric_cols = ['Payment Amount', 'Purchase PPU', 'Item Quantity',
                'Item Subtotal', 'Item Shipping & Handling',
                'Item Promotion', 'Item Tax', 'Item Net Total']

central_stats = pd.DataFrame({
    'Mean': df[numeric_cols].mean().round(2),
    'Median': df[numeric_cols].median().round(2),
    'Mode': df[numeric_cols].apply(lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else np.nan).round(2)
}, index=numeric_cols)

print("MEASURES OF CENTRAL TENDENCY")
print(central_stats)
```

```python
# Variance and Standard Deviation
variance_stats = pd.DataFrame({
    'Variance': df[numeric_cols].var().round(2),
    'Standard Deviation': df[numeric_cols].std().round(2)
}, index=numeric_cols)
print("\nVARIANCE AND STANDARD DEVIATION")
print(variance_stats)
```

```python
# Skewness
skewness_stats = pd.DataFrame({
    'Skewness': df[numeric_cols].skew().round(2)}, index=numeric_cols)
print("\nSKEWNESS")
print(skewness_stats)

```

```python
# sknewness visual
skew_values = df[numeric_cols].skew().round(2)
fig = px.bar(x=skew_values.index, y=skew_values.values,
             title="Skewness Heatmap - HIGHLY RIGHT-SKEWED SPEND DATA",
             labels={'x':'Metric', 'y':'Skewness (>0 = Right Tail)'},
             color=skew_values.values, color_continuous_scale='RdYlGn_r',
             height=500)
fig.update_layout(width=1000, height=500, autosize=False)
fig.show()
```

###Intrepretation
1. Right-Skewed Data (Most Metrics): Indicates many small values and fewer large values (outliers). This is typical for financial data where most transactions are small, but some are very large.
2. Negative Skewness (Item Promotion): Suggests many small discounts (or no discount) and a few very significant discounts.
3. Overall: The data is not normally distributed, highlighting that spending is concentrated at lower amounts with occasional high-value transactions.

```python
# Kurtosis
kurtosis_stats = pd.DataFrame({
    'Kurtosis': df[numeric_cols].kurtosis().round(2)}, index=numeric_cols)
print("\nKURTOSIS")
print(kurtosis_stats)
```

```python
# kurtosis visuals
kurtosis_values = df[numeric_cols].kurtosis().round(2)
fig = px.bar(x=kurtosis_values.index, y=kurtosis_values.values,
             title="Kurtosis",
             labels={'x':'Metric', 'y':'Kurtosis'},
             color=kurtosis_values.values, color_continuous_scale='RdYlGn_r',
             height=500)
fig.update_layout(width=1000, height=500, autosize=False)
fig.show()
```

### Interpretation


1. Positive Kurtosis (Leptokurtic): Most numerical columns show high positive kurtosis, indicating distributions with many extreme outliers. This is typical for financial data with infrequent, very large transactions.
2. Negative Kurtosis (Platykurtic): Item Promotion has negative kurtosis, meaning its distribution has lighter tails and fewer extreme outliers. This suggests promotional values are more clustered around their mean, with less variability in extreme discounts.
3. Overall: High positive kurtosis across most spending metrics confirms the presence of substantial outliers, especially on the higher end of expenditure, which is characteristic of real-world financial data.

## Bivariate Analysis

```python
# Scattar plot for payment amount through item quantity
plt.figure(figsize=(10, 6))
plt.scatter(df['Item Quantity'], df['Payment Amount'])
plt.title("Payment Amount vs Item Quantity")
plt.xlabel("Item Quantity")
plt.ylabel("Payment Amount")
plt.show()
```

### Interpretation
*   This scatter plot visualizes the relationship between the number of items purchased (`Item Quantity`) and the total payment for those items (`Payment Amount`).
*   Each point on the plot represents a single order, with its position indicating the quantity of items and the corresponding payment amount.
*   If points are clustered along an upward trend, it suggests a positive correlation: as the item quantity increases, the payment amount generally increases.
*   Outliers, points far from the main cluster, could indicate unusually high-value individual items or exceptionally large bulk purchases, offering insights into procurement anomalies.

```python
# Bar chart for average payment by product category
avg_payment_category = df.groupby('Product Category')['Payment Amount'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(avg_payment_category.index, avg_payment_category.values,color= Best_colors)
plt.xticks(rotation=90)
plt.title("Average Payment by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Average Payment")
plt.show()
```

### Interpretation
*   This bar chart displays the average payment amount for different product categories.
*   Taller bars indicate product categories that, on average, involve higher spending per transaction.
*   This helps identify which types of products, despite their frequency, command the highest average monetary value in procurement, suggesting areas of significant financial outlay.

```python
# Line chart for Monthly spend trend
df['Order Month'] = df['Order Date'].dt.to_period('M')

monthly_spend = df.groupby('Order Month')['Payment Amount'].sum()

plt.figure(figsize=(10, 6))
plt.plot(monthly_spend.index.astype(str), monthly_spend.values)
plt.xticks(rotation=45)
plt.title("Monthly Spending Trend")
plt.xlabel("Month")
plt.ylabel("Total Payment Amount")
plt.show()
```

### Interpretation
*   This line chart illustrates how total spending changes over time on a monthly basis.
*   The x-axis represents the months, and the y-axis shows the total payment amount.
*   By observing the line, we can identify trends such as increasing, decreasing, or stable spending patterns across the recorded months.
*   Peaks or dips in the line can highlight specific months with unusually high or low expenditure, which could indicate seasonal procurement, large one-off purchases, or budget adjustments.

```python
# Scattar plot for item quantity for shipping cost
df['Item Quantity'] = pd.to_numeric(df['Item Quantity'], errors='coerce')
df['Shipping Cost'] = pd.to_numeric(df['Item Shipping & Handling'], errors='coerce')

# Remove missing values
plot_df = df.dropna(subset=['Item Quantity', 'Item Shipping & Handling'])

# Plot
plt.figure(figsize=(8,6))

plt.scatter(plot_df['Item Quantity'],
            plot_df['Item Shipping & Handling'],
            alpha=0.6)

plt.title("Item Quantity vs Shipping Cost", weight='bold')
plt.xlabel("Item Quantity")
plt.ylabel("Item Shipping & Handling")

plt.tight_layout()
plt.show()
```

### Interpretation
*   This scatter plot explores the relationship between the `Item Quantity` and the `Item Shipping & Handling` cost.
*   Each point represents an order, showing how many items were purchased and their associated shipping cost.
*   Observing the plot helps determine if there's a direct correlation (e.g., more items generally lead to higher shipping costs) or if shipping costs are relatively flat regardless of quantity, possibly due to fixed-rate shipping or specific product types.
*   Outliers could highlight instances of unusually high or low shipping costs for certain item quantities, which might warrant further investigation.

```python
# Bar chart for Total spending per agency
agency_spend = df.groupby('Agency Name')['Item Net Total'].sum()

# Take top 10 agencies
top_agencies = agency_spend.sort_values(ascending=False).head(10)

plt.figure(figsize=(11,6))
plt.barh(top_agencies.index, top_agencies.values,color=Best_colors)
plt.title("Top 5 Agencies - Spending Share (FY26)", weight='bold')
plt.xlabel("Total Spending (USD)")
plt.ylabel("Agency Name")
plt.gca().invert_yaxis()
for index, value in enumerate(top_agencies.values):
    plt.text(value, index, f' ${value:,.0f}', va='center')

plt.tight_layout()
plt.show()
```

### Interpretation
*   This horizontal bar chart displays the total spending for each state agency.
*   Longer bars indicate agencies with higher overall expenditure, highlighting the top-spending departments.
*   This visualization helps to quickly identify which agencies are responsible for the largest portions of the total public expenditure, which can be useful for budget allocation reviews or identifying potential areas for cost efficiency.

## Multivariate Analysis

```python
# Bubble chart for product category which the spend money on each items
top_categories = df['Product Category'].value_counts().head(5).index
filtered = df[df['Product Category'].isin(top_categories)]

plt.figure()
plt.scatter(filtered['Item Quantity'],
            filtered['Payment Amount'],
            s=filtered['Item Quantity']*10)

plt.title("Quantity vs Payment (Top Categories)")
plt.xlabel("Item Quantity")
plt.ylabel("Payment Amount")
plt.show()
```

### Interpretation
*   This bubble chart visualizes the relationship between `Item Quantity` and `Payment Amount` for the top product categories, with the size of the bubbles representing a third variable (often `Item Quantity` itself, or another related metric to add depth).
*   Each bubble corresponds to an order within these categories.
*   Clustering of larger bubbles in the upper-right area would indicate categories where high quantities of items are purchased, leading to high payment amounts.
*   This visualization helps identify specific product categories that involve both large quantities and significant expenditure, and can reveal if certain categories drive high costs through volume or unit price.

```python
# Multi line chart fir Monthly Spend by Top 5 Categories
top5 = df['Product Category'].value_counts().head(5).index
df_top5 = df[df['Product Category'].isin(top5)]

monthly_cat = df_top5.groupby(
    [df_top5['Order Date'].dt.to_period('M'),
     'Product Category']
)['Payment Amount'].sum().unstack()

plt.figure(figsize=(10, 6))
for col in monthly_cat.columns:
    plt.plot(monthly_cat.index.astype(str), monthly_cat[col], label=col)

plt.xticks(rotation=45)
plt.title("Monthly Spending by Top Categories")
plt.xlabel("Month")
plt.ylabel("Payment Amount")
plt.legend()
```

### Interpretation
*   This multi-line chart visualizes the monthly spending trends for the top 5 product categories.
*   Each line represents a different product category, showing how its total payment amount changes over the months.
*   This allows for a direct comparison of spending patterns across different categories, identifying which categories are consistently high-spending, which show fluctuations, and which might have specific seasonal peaks or dips.
*   It helps to understand the relative contribution of each top category to the overall monthly expenditure.

### Interpretation
*   This heatmap displays the correlation coefficients between all pairs of numerical variables in the dataset.
*   The color intensity and direction (e.g., warmer colors for positive, cooler colors for negative) indicate the strength and type of relationship between variables.
*   Values close to +1 indicate a strong positive correlation, meaning variables tend to increase or decrease together.
*   Values close to -1 indicate a strong negative correlation, meaning as one variable increases, the other tends to decrease.
*   Values close to 0 indicate a weak or no linear correlation.
*   This visualization helps identify which variables move together and which are independent, providing insights for further analysis or feature selection.
